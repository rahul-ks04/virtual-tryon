```python
from PIL import Image
import matplotlib.pyplot as plt

person_img = Image.open(f"{DATA_ROOT}/train/image/00001_00.jpg")
cloth_img = Image.open(f"{DATA_ROOT}/train/cloth/00001_00.jpg")

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(person_img)
plt.title("Person")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cloth_img)
plt.title("Garment")
plt.axis("off")

plt.show()
```
